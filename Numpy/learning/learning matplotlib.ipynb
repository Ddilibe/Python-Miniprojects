```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, 'r*--')
plt.savefig('weatherdot.jpeg', bbox_inches='tight')
```

```python
plt.plot(x,y, color="green", linewidth=2, linestyle="dotted", markersize=10, marker="+")
plt.savefig('weatherplus.jpeg', bbox_inches='tight')
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = []
for i in range(len(max_t)):
    avg_t.append(int((max_t[i] + min_t[i])/2))
print(max_t, min_t, avg_t)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

# Loc is used to position the legend
plt.legend(loc="best", shadow=True, fontsize='large')

# The grid function is used to display the grid
plt.grid()
plt.savefig('weathergird.jpeg', bbox_inches='tight')
```

```python
company = ['Google', 'Amazon', 'PSFT', 'Facebook']
revenue = [80, 156, 87, 17]
profit = [40, 2, 34, 12]
```

```python
lab = np.arange(len(company))
plt.xticks(lab, company)
plt.ylabel('Revenue ($bil)')
plt.title('US Tech Stocks')
plt.xlabel("Companies")
plt.bar(lab-0.2, revenue, label="Revenue", width=0.4)
plt.bar(lab+0.2, profit, label="Profit", width=0.4)
plt.legend()
plt.savefig('barv.jpeg', bbox_inches='tight')
```

```python
lab = np.arange(len(company))
plt.yticks(lab, company)
plt.xlabel('Revenue ($bil)')
plt.title('US Tech Stocks')
plt.ylabel("Companies")
plt.barh(lab-0.2, revenue, label="Revenue", height=0.4)
plt.barh(lab+0.2, profit, label="Profit", height=0.4)
plt.legend()
plt.savefig('barh.jpeg', bbox_inches="tight")
```

# Working In a Way With Pandas

```python
data = {
    'patient':[
        'Bob','Robert','Kusum','Ahmed','Mike','John','Ketan',
        'Adbul','Maria','Nikisha','Veeral','Dhaval','Sara'
    ],
    'Blood Sugar': [113,85,90,150,149,88,93,115,135,80,77,82,129]
}
print(data)
```

```python
new_data = pd.DataFrame(data)
new_data.set_index('patient', inplace=True)
print(new_data)
```

```python
new_data.rename(
    columns={
        'Blood Sugar':'blood_sugar'
    }, inplace=True
)
```

```python
((new_data['blood_sugar'] > 80) & (new_data['blood_sugar'] <100)).sum()
```

# Visualizing Data in a PieChart

```python
data_pie = {
    'home_rent':1400,
    'food':600,
    'phone':300,
    'car':410,
    'Other Utilities':250
}
data_pie
```

```python
exp_vals = data_pie.values()
exp_labels = data_pie.keys()
```

```python
plt.pie(exp_vals, labels=exp_labels,autopct='%0.0F%%', 
        shadow=True, explode=[0,0.2,0.3,0,1],
       startangle=180)
plt.axis('equal')
plt.show
```
