```python
import faker
import pandas as pd
```

```python
def random_name(number):
    new_name = []
    name = faker.Faker()
    for i in range(number):
        new_name.append(name.name())
    return (new_name)
```

```python
def random_generate(a, b, number):
    import random
    new_list = []
    for i in range(number):
        new_list.append(random.randint(a, b))
    return (new_list)
```

```python
def get_dataset(a, b, number):
    name = random_name(number)
    age = random_generate(11, 20, number)
    score = random_generate(0, 100, number)
    return name, age, score
```

```python
name, age, score = get_dataset(1,1, 300)
```

```python
new_data = {'name':name, 'age':age, 'score':score}
new_data = pd.DataFrame(new_data)
```
