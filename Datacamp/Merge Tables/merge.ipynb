```python
import faker
import pandas as pd
```

```python
dir(faker.Faker())
```

```python
first_name, last_name, email, phone_number, street_name, id, url, state, job = [], [], [], [], [], [], [], [], []
f = faker.Faker()
for i in range(100):
    first_name.append(f.first_name())
    last_name.append(f.last_name())
    email.append(f.email())
    phone_number.append(f.phone_number())
    street_name.append(f.street_name())
    id.append(f.uuid4())
    url.append(f.url())
    state.append(f.state())
    job.append(f.job())
    
import random
for i in range(100):
    jam = random.choice([ id, url, state, job])
    we = random.randint(0, 99)
    jam[we] = None
    
```

```python
first_dictionary = {
    'id':id,
    'first_name':first_name,
    'last_name':last_name,
    'email':email,
    'phone_number':phone_number,
    'street_name':street_name
}
second_dictionary = {
    'id':id,
    'first_name':first_name,
    'last_name':last_name,
    'website_url':url,
    'state':state,
    'job':job
}
```

```python
df1 = pd.DataFrame(first_dictionary)
df2 = pd.DataFrame(second_dictionary)

print(df1.shape, df2.shape)
```
