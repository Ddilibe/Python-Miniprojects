```python
import numpy as np
from scipy import sparse
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
# Using K-Nearest Neighbor (KNN) Algorithm for a Classification Problem
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
cancer = load_breast_cancer()
```

```python
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)
training_accuracy, test_accuracy, neighbors_settings = [], [], range(1, 11)
for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(x_train, y_train)
    training_accuracy.append(clf.score(x_train, y_train))
    test_accuracy.append(clf.score(x_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression, Ridge
```

```python
lr = LinearRegression().fit(x_train, y_train)
print(f"lr.coef_: {lr.coef_}\nlr.intercept_: {lr.intercept_}")
print(f"Training set score: {lr.score(x_train, y_train)}")
print(f"Test set score: {lr.score(x_test, y_test)}")
rd = Ridge().fit(x_train, y_train)
print(f"rd.coef_: {rd.coef_}\nrd.intercept_: {rd.intercept_}")
print(f"Training set score using Ridge: {rd.score(x_train, y_train)}")
print(f"Test Set Score Using Ridge: {rd.score(x_test, y_test)}")
# Ridge Regression improve the accuracy to predicting test data but reduces the accuracy to predicting training data
# This is compared to LinearRegression Class
```
